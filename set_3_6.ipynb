```python
# write a function to display the deteails of the employee
# name, dept, salary
# call it using the positional arguments as well as
# kwyword arguments..

def emp_details(name, dept, salary):
  return f'Name of the emp ->{name}. salary ->{salary}. dept ->{dept}.'

# positional Arguments
emp_details('prAJJAL', 'Software Enginerring', 60)

# Keyword
emp_details(name = 'prAJJAL', dept = 'SE', salary = 60)
```
