```python
class House:
  color = 'Blue'
  room = 5

  # constructor is automatically called when the object is create
  def __init__(self, user_color):
     print('Init function is called !')
     print(self)
     self.house_color = user_color
     print(self)
  def display(self):
     print(self.color)
     print(self.room)

house1 = House('Green')
house2 = House('Red')
print(house1.house_color)
print(house2.house_color)
print(house1.room)
print(house2.room)
```

```python
class Student:
  # this is class attributes -> which is remains for
  # all the objects or the instance
  college = 'IIT'
  def __init__(self, name, locTION):
    # Instance attributes which is defind into the defind
    # into the constructor method
    self.student_name = name
    self.student_location = locTION
  def display(self):
      print(f'Student Name: {self.student_name}')
      print(f'Student Location: {self.student_location}')
      print(f'College Name: {self.college}')

s1 = Student('Harshita','Mumbai')
s2 = Student('Raj','Delhi')
s1.display()
s2.display()
```

```python
# Create a class of employee
# class attributes company
# Employee detail -> Name, Location , department, salary
# make a method to display all the details of the employee

class Employee:
    company = "Google"
    def __init__(self, name, location, department, salary):
        self.name = name
        self.location = location
        self.department = department
        self.salary = salary

    def display(self):
          print("Name: ", self.name)
          print("Location: ", self.location)
          print("Department: ", self.department)
          print("Salary: ", self.salary)
          print("Company: ", self.company)

employee1 = Employee("Jay", "New York", "IT", 50000)
employee1.display()
employee2 = Employee("Ram", "Dubai", "HR", 55000)
employee2.display()
```

```python
# create a class of book with instance attribute
# title, writer, published_on
# display the deteails with the display method

class Book:
    def __init__(self, title, writer, published_on):
        self.title = title
        self.writer = writer
        self.published_on = published_on
    def display(self):
      print("Title: ", self.title)
      print("Writer: ", self.writer)
      print("Published On: ", self.published_on)

book1 = Book("The Alchemist", "Paulo Coelho", "1988")
book1.display()
book2 = Book("The Alchemist", "Paulo Coelho", "1988")
book2.display()
```

```python
# Inheritance -> Is a one of it's pillar which
# inherit the porperty from the parent class to the child class
# This is parent class
class Animal:
    def sound(self):
      print("Animal makes a sound")

# This is inherited class which is inherited by
# animal class
class Dog(Animal):
   pass
dog = Dog()
dog.sound()
```
