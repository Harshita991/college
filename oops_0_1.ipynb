```python
# why we use the object oriented programing ?
# The problem with procedural programing that we can't
# into the program-> ex- you have to remeber that..
# so, that's why we use the oops.

# object oriented programing
# we make the blueprint of any object or entity which is know as class
# the real world entity we use into that class
# that is object.
```

```python
# Here we would make a student class
# here we have created class which has attribute name
class Student:
  # By default harshita
  # class attributes -> this class attribute is same
  # for all the objects or instances
  name = 'Harshita'

  # counstructor is automatically called when we created
  # an object.. by giving one property or argument
  def __init__(self, location):
    self.attr_location = location

# here we have created a student object
obj1 = Student('jaipur')

# here we have accessed the object's attributes
print(obj1.attr_location)
obj2 = Student('delhi')
print(obj2.attr_location)
```

```python
# make a class of teacher, experience, name, domain
class teacher:
    def __init__(self, name, experience, domain):
      self.name = name
      self.experience = experience
      self.domain = domain
      print(self.name + ' ' + str(self.experience) + ' ' + self.domain)

obj1 = teacher('Harshita', 1, 'python')
obj2 = teacher('Neha', 2, 'java')
obj3 = teacher('Arpita', 3, 'c++')
obj4 = teacher('SHYAM', 4, 'c')
obj5 = teacher('Anjali', 5, 'c#')

```
