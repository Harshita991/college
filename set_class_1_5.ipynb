```python
s = set()
print("s element")
for i in range(5):
  value = int(input("enter element"))
  s.add(value)
print(s)
```

```python
# use while loop
# brute force approach
s = set()
n = 5
while n > 0:
  element = int(input('enter a element:'))
  s.add(element)
  n-= 1

```

```python
# 3remove elements: use (remove)method
s = {1,2,3,4,5}
s.remove(2)
print(s)
```

```python
# remove elements: use (remove) method
# discard is used to remove the particular element from the
# set if the element is preesent into the set, then it will
# remove that element . otherwise it will not throw any error
s = {1,2,3,4,5}
result = s.discard(50)
print(result)
```

```python
# 4 pop method: remove any random element from the set and return that.
s = {10,20,30,40}
remove = s.pop()
print("Removed element:", remove)
print("Remaining set:", s)
```
