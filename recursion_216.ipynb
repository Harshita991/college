```python
# write program to handel the error which is raised dictionry
# key.. the key which doesn't exist we want to acess it
# try, except, else, finally
my_dict = {
    "name": "Harshita",
    "age": 21,
    "city": "Delhi"
}
key = input("Enter the key to access: ")

try:
    value = my_dict[key]   # KeyError may occur
except KeyError:
    print("Error: The key does not exist in the dictionary.")
else:
    print("Value:", value)
finally:
    print("Dictionary access operation completed.")
```
