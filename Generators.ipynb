```python
# Generators -> in python generators are special function or expression
# used to create iterators that produce a sequence of value one at time,
# on demand (lazy evalution), insted of building and stroing
# the entire sequence in memory

# qus .1
# def ele():
#   return [1, 2, 3]

def ele():
  yield 1
  yield 2
  yield 3

generator = ele()
print(next(generator))
print(next(generator))
print(next(generator))

```

```python
# qus 2. print some 10 the prime numbers in range of 1 to 100000 using generators

def gen_prime():
  for num in range(2, 100000):
     is_prime = True
     for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
     if is_prime:
        yield num

g = gen_prime()

print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))

```

```python
# qus.3 write a program to print odd number win range 1 to 100000 using generator

def gen():
  for i in range(1, 100000):
    if i % 2 != 0:
      yield i
g = gen()
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```
