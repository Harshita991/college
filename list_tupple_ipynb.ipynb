```python
li = [[1,2,3], [4,5,6], [7,8,9]]
transpose_mat = []
for i in range(len(li)):
  row = []
  for j in range(len(li[i])):
     row.append(li[j][i])
  transpose_mat.append(row)
print(transpose_mat)
```

```python
# find the sum of all the elements of the given tuple(1,2,3,4,5,6,7,8)

t1 = (1,2,3,4,5,6,7,8)
sum = 0
for i in t1:
  sum += i
print(sum)
```

```python
# qus . 2 convert a tuple of string into the one string ->
# ('r', 'a', 'h', 'u', 'l')-> 'rahul'

t = ('r', 'a', 'h', 'u', 'l')
name = ''
for ch in t:
  name += ch
print(name)
```
