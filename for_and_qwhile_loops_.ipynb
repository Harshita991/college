```python

# qus. 7

count = 0
num = int(input("Enter a number: "))

for i in range(1, num + 1):
    if num % i == 0:
        count += 1

if count == 2:
    print(f"{num} is a prime number.")
```

```python
# qus.8
num = input("Enter a number: ")

count = 0
for digit in num:
    if digit.isdigit():   # counts digits only
        count += 1

print("Number of digits:", count)
```

```python
# qus.9

num = input("Enter a number: ")

total = 0
for digit in num:
    if digit.isdigit():     # ensures only digits are summed
        total += int(digit)

print("Sum of digits:", total)
```

```python
# qus.10

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# The LCM will be at least the maximum of the two numbers
max_num = max(num1, num2)

for i in range(max_num, (num1 * num2) + 1):
    if i % num1 == 0 and i % num2 == 0:
        lcm = i
        break

print("LCM of", num1, "and", num2, "is:", lcm)
```

```python
# qus. 2

num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed number:", rev)
```

```python
# qus.3

num = int(input("Enter a number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

if num == rev:
    print("Palindrome number")
else:
    print("Not a palindrome")
```

```python
# qus 4

num = int(input("Enter a number (0 to stop): "))

while num != 0:
    print("You entered:", num)
    num = int(input("Enter a number (0 to stop): "))

print("Program ended.")
```

```python
# qus.5

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b != 0:
    temp = b
    b = a % b
    a = temp

print("GCD (HCF) =", a)
```

```python
# qus.6

start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))

# Make sure we start from the next number after 'start'
i = start

while i <= end:
    if i % 2 != 0:
        print(i)
    i += 1

```
