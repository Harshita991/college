```python
# Decorators
# A Python decorators are a powerful feature that allows a programmer to modify or
# extend the behavior of a function or a class without changing its source code
# They provide a reusable way to wrap an existing function with additional functionality,
# using the @ symbol for simple applicatio
```

```python
# Decorators
# qus.1
def add_fun(fun):
  def wrapper():
    print('This is the wrapper function')
    fun()

def hello():
  print('Hello from the hello function')
hello()
```

```python
# qus.2
def add_fun(fun):
  def add_func():
     print('This is the msg!')
     fun()
  return add_func

@add_fun
def hello():
  print('Hello from the hello function')
hello()
```

```python
# qus.3
def my_decorator(gift):
  def wrapper():
    print('Adding some details into the book!')
    gift()
    print('Adding some feature after calling the book function')
  return wrapper

@my_decorator

def gift():
  print('This is the gift')
gift()
```

```python
# qus 5. create a decorator which check if the number is even before calling the function

# This is the decorator in which we have passed into the display run
def check_even(fun):
  # so, here the number which we have passed into the display num = 5
  def func(num):
    # Before caaling the function we are checking that the num is even or odd
    if num%2==0:
      print('Even')
      # calling the original function -> display(num = 5)
      fun(num)
    else:
      print('Odd')
  return func
  # here we have added a decorator which takes n, and display function as an argument
@check_even

def display(n):
  print(f'The number is:{n}')
# 1. calling the display dunction by passing the num = 5
display(5)

```

```python
# qus 6.
is_permission = False

def permission_access(fun):
  def wrapper():
    if is_permission:
      fun()
    else:
      print('Permission Denied!')
  return wrapper
@permission_access
def data():
  print('This is the secret data!')
data()

```
